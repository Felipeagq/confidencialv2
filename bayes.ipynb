```python
def clean_data(texto):
    # Minuscula
    texto = texto.lower()
    
    # Eliminamos el https
    if 'http' in texto:
        end = texto.index('http')
        texto = texto[:end]
    
    # reemplazamos por "nada" los signos de puntuación
    punctuation = '!¿?¡"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~'
    for punct in punctuation:
        texto = texto.replace(punct,'')
        
        
    # reemplazamos por "nada" los signos de puntuación
    acentos = 'áàéêíóú'
    acentos_no = 'aaeeiou'
    for a,b in zip(acentos,acentos_no):
        texto = texto.replace(a,b)
    
    # Separamos el texto -> lista
    texto = texto.split()
    
    # Reemplazamos cualquier espacio sobrante por "nada"
    for i in range(len(texto)):
        texto[i] = texto[i].replace(' ','')
    
    # Si mi elemento de la lista es un numero, lo elimino       
    for palabra in texto:
        if palabra.isdigit():
            texto.remove(palabra)

    # si tengo palabras combinadas con numero, elimino los numeros
    for i in range(len(texto)):
        s = texto[i]
        texto[i] = ''.join([i for i in s if not i.isdigit()])
        
    # elimino las palabras con 2 o menos caracteres
    to_remove = []
    for i in range(len(texto)):
        s = texto[i]
        if len(s) <= 3:
            to_remove.append(s)
    for remove in to_remove:
        texto.remove(remove)
            
            
    return texto
```

```python
data['text'] = data['text'].apply(lambda x: clean_data(x))
```
