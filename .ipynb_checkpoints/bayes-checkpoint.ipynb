```python
data = pd.read_csv('tuits_bayes.csv', encoding='latin1')
```

```python
data.drop('status_id',axis=1,inplace=True)
```

```python
# LIMPIEZA DE LOS DATOS 
def clean_data(texto):
    # Minuscula
    texto = texto.lower()
    
    # Eliminamos el https
    if 'http' in texto:
        end = texto.index('http')
        texto = texto[:end]
    
    # reemplazamos por "nada" los signos de puntuación
    punctuation = '!¿?¡"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~'
    for punct in punctuation:
        texto = texto.replace(punct,'')
        
        
    # reemplazamos por "nada" los signos de puntuación
    acentos = 'áàéêíóú'
    acentos_no = 'aaeeiou'
    for a,b in zip(acentos,acentos_no):
        texto = texto.replace(a,b)
    
    # Separamos el texto -> lista
    texto = texto.split()
    
    # Reemplazamos cualquier espacio sobrante por "nada"
    for i in range(len(texto)):
        texto[i] = texto[i].replace(' ','')
    
    # Si mi elemento de la lista es un numero, lo elimino       
    for palabra in texto:
        if palabra.isdigit():
            texto.remove(palabra)

    # si tengo palabras combinadas con numero, elimino los numeros
    for i in range(len(texto)):
        s = texto[i]
        texto[i] = ''.join([i for i in s if not i.isdigit()])
        
    # elimino las palabras con 2 o menos caracteres
    to_remove = []
    for i in range(len(texto)):
        s = texto[i]
        if len(s) <= 3:
            to_remove.append(s)
    for remove in to_remove:
        texto.remove(remove)
            
            
    return texto
```

```python
data['text'] = data['text'].apply(lambda x: clean_data(x))
```

```python
# FILTRO DE USUARIOS, QUEDANDO SOLO LOS QUE TIENEN MÁS DE 3 TWEETS

nombres = data['screen_name'].unique()
conteo = []
for nombre in nombres:
    cantidad = data[data['screen_name']==nombre]
    if len(cantidad) <=5:
        conteo.append(nombre)
        

los_que_quedan = []
for nombre in nombres:
    if nombre not in conteo:
        los_que_quedan.append(nombre)
        
los_que_quedan
```

```python
lopez = data[data['screen_name']=='lopezobrador_']
MSFT = data[data['screen_name']=='MSFTMexico']
UNAM = data[data['screen_name']=='UNAM_MX']
CMLL = data[data['screen_name']=='CMLL_OFICIAL']
```

```python
# PROBABILIDAD APRIORI

n = len(lopez) + len(MSFT) + len(UNAM) + len(CMLL)
prob_lopez = len(lopez)/n
prob_MSFT = len(MSFT)/n
prob_UNAM = len(UNAM)/n
prob_CMLL = len(CMLL)/n
print(prob_lopez)
print(prob_MSFT)
print(prob_UNAM)
print(prob_CMLL)
```

```python
# Hallamos las palabras unicas de por cada usuario

def palabras_de(dataframe):
    conjunto=set()
    for i in range(len(dataframe)):
            tweet = dataframe.iloc[i]['text']
            for palabra in tweet:
                conjunto.add(palabra)
    return conjunto

palabras_lopez = palabras_de(lopez)
palabras_UNAM = palabras_de(UNAM)
palabras_CMLL = palabras_de(CMLL)
palabras_MSFT = palabras_de(MSFT)
```

```python
# Creamos diccionarios

dict_lopez = dic = dict.fromkeys(palabras_lopez,1)
dict_UNAM = dic = dict.fromkeys(palabras_UNAM,1)
dict_CMLL = dic = dict.fromkeys(palabras_CMLL,1)
dict_MSFT = dic = dict.fromkeys(palabras_MSFT,1)

"""dict_lopez = dic = dict.fromkeys(palabras_lopez,0)
dict_UNAM = dic = dict.fromkeys(palabras_UNAM,0)
dict_CMLL = dic = dict.fromkeys(palabras_CMLL,0)
dict_MSFT = dic = dict.fromkeys(palabras_MSFT,0)"""


# Creamos el diccionario que será nuestra "bolsa" de palabras

bolsa = {}
bolsa.update(dict_lopez)
bolsa.update(dict_UNAM)
bolsa.update(dict_CMLL)
bolsa.update(dict_MSFT)
```

```python
# Conteo de las palabras unicas

def dict_words(dataframe, dic):
    for i in range(len(dataframe)):
            tweet = dataframe.iloc[i]['text']
            for palabra in tweet:
                dic[palabra] = dic[palabra] + 1 
    return dic

dict_lopez = dict_words(lopez,dict_lopez )
dict_UNAM = dict_words(UNAM,dict_UNAM )
dict_CMLL = dict_words(CMLL,dict_CMLL )
dict_MSFT = dict_words(MSFT,dict_MSFT )

```

```python
for palabra in bolsa.keys():
    dict_lopez.setdefault(palabra,1)
    dict_UNAM.setdefault(palabra,1)
    dict_CMLL.setdefault(palabra,1)
    dict_MSFT.setdefault(palabra,1)
```

```python
# palabras_prob_lopez 
def palabra_prob(dic):
    total_palabras_usuario = sum(dic.values())
    for palabra in dic.keys():
        dic[palabra] = dic[palabra]/total_palabras_usuario
    return dic

dict_lopez = palabra_prob(dict_lopez)
dict_UNAM = palabra_prob(dict_UNAM)
dict_CMLL = palabra_prob(dict_CMLL)
dict_MSFT = palabra_prob(dict_MSFT)
```

```python

def who(mensaje):
    orden = (('lopez',dict_lopez,prob_lopez),
            ('UNAM',dict_UNAM,prob_UNAM),
            ('CMLL',dict_CMLL,prob_CMLL),
            ('MSFT',dict_MSFT,prob_MSFT))

    mensaje = mensaje.replace(',','')
    mensaje = mensaje.split()
    resp= []
    
    for i in range(len(orden)):
        usuario,d,p = orden[i]
        prob_cond = 1
        porb_posteriori = 1
        for palabra in mensaje:
            prob_cond = prob_cond * (d[palabra] * p)
            acum = (prob_UNAM*dict_UNAM[palabra]) + (prob_CMLL*dict_CMLL[palabra])+(prob_lopez*dict_lopez[palabra]) + (prob_MSFT*dict_MSFT[palabra] )      
            #print('acum',acum)
            porb_posteriori = porb_posteriori * acum
        res = (prob_cond * p)/porb_posteriori
        resp +=[res]
        print(usuario,'\t',res)


        grande = max(resp)
        n = resp.index(grande)
        user = orden[n][0]
    print(f' \n el correo fu escrito por {user}')
```
