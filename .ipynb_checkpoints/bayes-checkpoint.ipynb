```python
data = pd.read_csv('tuits_bayes.csv', encoding='ISO-8859-1')
```

```python
data
```

```python
data.drop('status_id',axis=1,inplace=True)
```

```python
def clean_data(texto):
    texto = texto.lower()
    if 'http' in texto:
        end = texto.index('http')
        texto = texto[:end]
        
    texto = texto.replace(',','').replace(':','').replace('?','').replace('¿','').replace('"','').replace('.','').split()
    
    for i in range(len(texto)):
        texto[i] = texto[i].replace(' ','')
        
    for i in range(len(texto)):
        s = texto[i]
        if len(s) <= 2:
            texto.remove(s)
            
    for palabra in texto:
        if palabra.isdigit():
            texto.remove(palabra)

    for i in range(len(texto)):
        s = texto[i]
        texto[i] = ''.join([i for i in s if not i.isdigit()])
        

            
    return texto
```

```python
data['text'] = data['text'].apply(lambda x: clean_data(x))
```
